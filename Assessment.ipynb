```python
import numpy as np
if not hasattr(np, 'bool8'):
    np.bool8 = np.bool_

# --- Step 2: Imports ---
import pandas as pd
import seaborn as sns
import plotly.express as px
import cufflinks as cf
from plotly.offline import iplot

# --- Step 3: Go offline for interactive plotting ---
cf.go_offline()
df = pd.read_csv("C:\\Users\dell\Downloads\\covid_india_data.csv")
print(df.head())
checknull=df.isnull().sum()
print(checknull)
print(df.dtypes)

print("Basic Information:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())

print("\nMaximum new cases reported in a single day:", df['new_cases'].max())
print("Average new cases per day:", df['new_cases'].mean())
print("Total deaths reported:", df['new_deaths'].sum())
#time trend graph
df['date']=pd.to_datetime(df['date'])
sns.lineplot(x='date', y='new_cases', data=df)
fig = px.line(df, x='date', y='new_cases', title='date vs. new cases')
fig.show()

sns.lineplot(x='date',y='new_deaths',data=df)
fig=px.line(df,x='date',y='new_deaths',title='date vs. new deaths')
fig.show()


sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')



#realtionship graphs correlation between columns
sns.scatterplot(x='new_cases', y='new_deaths',color="orange", data=df)
fig = px.scatter(df, x='new_cases', y='new_deaths', title='new_cases vs. new_deaths')
fig.show()

#distribution graphs
sns.histplot(df['new_cases'], bins=5, kde=True)
fig=px.histogram(df,x='new_cases',nbins=30)
fig.show()

fig = px.pie(df, names=['new_cases', 'new_deaths'], values=[df['new_cases'].sum(), df['new_deaths'].sum()],
             title='Cases vs Deaths Distribution')
fig.show()
```
